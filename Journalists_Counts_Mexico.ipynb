```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point, Polygon
import geojsonio
%matplotlib inline
import matplotlib as plt
import folium
from IPython.display import HTML, display
import geopy
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_journalists_and_media_workers_killed_in_Mexico'
tables = pd.read_html(link)
TOIs = pd.concat(tables[1:2])
TOIs.columns = ['Date', 'Name','Location','Media Outlet', 'Statuas','Refs']
##TOIs = TOIs[TOIs.Country != 'Country']
TOIs

```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
TOIs['City_Coord'] = TOIs['Location'].apply(geolocator.geocode)
TOIs
```
