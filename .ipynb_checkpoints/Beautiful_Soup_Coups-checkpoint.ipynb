```python
from requests import get
from bs4 import BeautifulSoup
import re
url = 'https://en.wikipedia.org/wiki/List_of_coups_d%27%C3%A9tat_and_coup_attempts_by_country'
```

```python
response = get(url)
print(response.text[:500])
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
print(html_soup)
```

```python
Cntry = html_soup.find_all('h2')
print(type(Cntry))
print(len(Cntry))
```

```python
F_Cntry = Cntry[1]
F_Cntry
```

```python
Country = F_Cntry.find('span')
print(Country.text)
```

```python
names = []

for container in Cntry:
    if container.find('span', class_ = 'mw-headline') is not None:
        name = container.text
        names.append(name)
        
        ##overthrower = container.find('li', class_ = 'mw-redirect').text
        ##overthrowers.append(overthrower)
        
##year = container.h3.find('span', class_ = 'lister-item-year').text
  ##      years.append(year)
  
```

```python
print(names)
```
