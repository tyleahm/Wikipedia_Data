```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point, Polygon
import geojsonio
%matplotlib inline
import matplotlib as plt
import folium
from IPython.display import HTML, display
from folium import plugins
##Research Cartopy
##https://towardsdatascience.com/around-the-world-with-anthony-bourdain-a-folium-tutorial-7e9ad63fb650
```

```python
link = "C://Users//fcmun//Documents//HTML_Test.html"
tables = pd.read_html(link)
TOIs = pd.concat(tables[2:11])
TOIs.columns = ['Country', 'Location','Coordinates','Operational date', 'Number of Rockets launches','Heaviest Rocket Launched','Highest achieved altitude', 'Notes']
TOIs = TOIs[TOIs.Country != 'Country']
TOIs

##https://en.wikipedia.org/wiki/List_of_rocket_launch_sites
```

```python
TOIs[['DMS', 'DD']] = TOIs['Coordinates'].str.split('/', n=1, expand=True)
TOIs[['DD1', 'DD2']] = TOIs['DD'].str.split('°S|°N', n=1, expand=True)
TOIs['DD1_Extra'] = pd.np.where(TOIs.DD.str.contains("S"), "-",'')
TOIs['DD1'] = TOIs['DD1_Extra'] + TOIs['DD1']
TOIs['DD2_Extra'] = pd.np.where(TOIs.DD.str.contains("W"), "-",'')
TOIs['DD2'] = TOIs['DD2_Extra'] + TOIs['DD2']
TOIs['DD1'] = TOIs.DD1.str.replace(r"[^0-9.-]",'')
TOIs['DD2'] = TOIs.DD2.str.replace(r"[^0-9.-]",'')
TOIs['DD1'] = pd.to_numeric(TOIs['DD1'], errors='coerce')
TOIs['DD2'] = pd.to_numeric(TOIs['DD2'], errors='coerce')
columns = ['DD', 'DD1_Extra', 'DD2_Extra']
TOIs = TOIs[TOIs['DD2'].notnull()]
TOIs = TOIs[TOIs['DD1'].notnull()]
TOIs.drop(columns, inplace=True, axis=1)
TOIs
```

```python
TOIs[['StartY', 'EndY']] = TOIs['Operational date'].str.split('–', n=1, expand = True)
TOIs
```

```python
map_hooray = folium.Map()
map_hooray
```

# Heat Map

```python
TOIs2 = TOIs[['DD1', 'DD2']]
TOIs2['Launches'] = TOIs['Number of Rockets launches']
TOIs2['Launches'] = TOIs2.Launches.str.replace(r"[^0-9.-]",'')
TOIs2['Launches'] = TOIs2.Launches.str.replace("-",'')
TOIs2 = TOIs2.dropna(axis=0, subset=['DD1','DD2', 'Launches'])
TOIs2['Launches'] = TOIs2['Launches'].astype(float)
##TOIs2.to_csv('float.csv')
```

```python
heat = [[row['DD1'],row['DD2']] for index, row in TOIs2.iterrows()]
```

```python
from folium.plugins import HeatMap
HeatMap(heat).add_to(map_hooray)
map_hooray
```

# Time Series Heat Map

```python
TOIs2 = TOIs[['DD1', 'DD2']]
TOIs2['lc'] = TOIs['Number of Rockets launches']
TOIs2['lc'] = TOIs2.lc.str.replace(r"[^0-9.-]",'')
TOIs2['lc'] = TOIs2.lc.str.replace("-",'')
TOIs2 = TOIs2.dropna(axis=0, subset=['DD1','DD2', 'lc'])
TOIs2 = TOIs2[TOIs2.lc  != '']
TOIs2['lc'] = TOIs2['lc'].astype(float)
TOIs2
```

```python
heat = [[[row['DD1'],row['DD2']] for index, row in TOIs2[TOIs2['lc']== i].iterrows()] for i in range (0,13)]
print(heat)
```

```python
hm = plugins.HeatMapWithTime(heat)
hm.add_to(map_hooray)
# Display the map
map_hooray
```
